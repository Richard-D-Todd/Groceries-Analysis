# Load emails into python#

```python
#import email module
import email
from requests_html import HTML
```

```python
#Open local file
a = open(r'C:\Users\Richard\Documents\test email.htm', 'r')

#Convert to string
raw_a = str(a.read())

b = email.message_from_string(raw_a)
body = ""

if b.is_multipart():
    for part in b.walk():
        ctype = part.get_content_type()
        cdispo = str(part.get('Content-Disposition'))
        
        #Skips ant text attachments
        if ctype == 'text/plain' and 'attachment' not in cdispo:
            body = part.get_payload(decode=True)
            break
            
#not multipart
else:
    body = b.get_payload(decode=True)
```

```python
#print the body of the email and close the email file
print(body)
a.close()
```

Now that the email body is extracted I need to parse the HTML text for the table

```python
#Grab the HTML from the body of the email
html = HTML(html=body)
```
